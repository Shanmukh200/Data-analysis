```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 22, 23, 24, 22],
    'Math_Score': [88, 92, 85, 95, 90],
    'Science_Score': [90, 85, 88, 92, 95],
    'English_Score': [85, 87, 89, 90, 86]
}

df = pd.DataFrame(data)
print(df)

```

```python
# Summary statistics
print(df.describe())
```

```python
# Mean scores
mean_scores = df[['Math_Score', 'Science_Score', 'English_Score']].mean()
print(mean_scores)
```

```python
# Highest and lowest scores
highest_scores = df[['Name', 'Math_Score', 'Science_Score', 'English_Score']].max()
lowest_scores = df[['Name', 'Math_Score', 'Science_Score', 'English_Score']].min()

print(f"Highest Scores:\n{highest_scores}")
print(f"Lowest Scores:\n{lowest_scores}")
```

```python
import matplotlib.pyplot as plt

# Plotting the average scores
mean_scores.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Average Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.show()
```

```python
# Plotting the age distribution
df['Age'].plot(kind='hist', bins=5, color='gray', edgecolor='black')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
