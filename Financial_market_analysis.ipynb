```python
#Needed to intall yfinance package additionally to import mpl finance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# Fetch historical stock data
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']  # Updated list of stocks
data = yf.download(tickers, start='2020-01-01', end='2023-01-01')['Adj Close']
print(data.head())
```

```python
# Time Series Analysis: Plot closing prices
plt.figure(figsize=(14, 7))
for ticker in tickers:
    plt.plot(data[ticker], label=ticker)
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.legend()
plt.show()
```

```python
# Moving Averages
def plot_moving_averages(stock_data, ticker):
    plt.figure(figsize=(14, 7))
    stock_data['20_SMA'] = stock_data['Adj Close'].rolling(window=20).mean()
    stock_data['50_SMA'] = stock_data['Adj Close'].rolling(window=50).mean()
    plt.plot(stock_data['Adj Close'], label='Adj Close')
    plt.plot(stock_data['20_SMA'], label='20 Day SMA')
    plt.plot(stock_data['50_SMA'], label='50 Day SMA')
    plt.title(f'{ticker} - Moving Averages')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.legend()
    plt.show()

for ticker in tickers:
    stock_data = yf.download(ticker, start='2020-01-01', end='2023-01-01')
    plot_moving_averages(stock_data, ticker)
```

```python
# Bollinger Bands
def plot_bollinger_bands(stock_data, ticker):
    plt.figure(figsize=(14, 7))
    stock_data['20_SMA'] = stock_data['Adj Close'].rolling(window=20).mean()
    stock_data['20_STD'] = stock_data['Adj Close'].rolling(window=20).std()
    stock_data['Upper_Band'] = stock_data['20_SMA'] + (2 * stock_data['20_STD'])
    stock_data['Lower_Band'] = stock_data['20_SMA'] - (2 * stock_data['20_STD'])
    plt.plot(stock_data['Adj Close'], label='Adj Close')
    plt.plot(stock_data['20_SMA'], label='20 Day SMA')
    plt.plot(stock_data['Upper_Band'], label='Upper Band')
    plt.plot(stock_data['Lower_Band'], label='Lower Band')
    plt.fill_between(stock_data.index, stock_data['Lower_Band'], stock_data['Upper_Band'], color='gray', alpha=0.3)
    plt.title(f'{ticker} - Bollinger Bands')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.legend()
    plt.show()

for ticker in tickers:
    stock_data = yf.download(ticker, start='2020-01-01', end='2023-01-01')
    plot_bollinger_bands(stock_data, ticker)
```

```python
# Candlestick Charts
import mplfinance as mpf

for ticker in tickers:
    stock_data = yf.download(ticker, start='2020-01-01', end='2023-01-01')
    stock_data = stock_data[['Open', 'High', 'Low', 'Close', 'Volume']]
    mpf.plot(stock_data, type='candle', style='charles', title=f'{ticker} - Candlestick Chart', volume=True)
```

```python
# Correlation Matrix
corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Stock Prices')
plt.show()
```
